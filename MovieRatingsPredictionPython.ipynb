```python
data_lm = TextLMDataBunch.from_df(".",train_df = imdb_out[["movie_title","storyline", "imdb_score"]][:len(imdb_out)-1000], 
                       valid_df = imdb_out[["movie_title","storyline", "imdb_score"]][len(imdb_out)-1000:],
                        text_cols="storyline", label_cols="imdb_score")
```

```python
learn = language_model_learner(data_lm, AWD_LSTM, drop_mult=0.5)
```

```python
learn.freeze_to(-2)
learn.fit_one_cycle(1, 1e-2)
```
